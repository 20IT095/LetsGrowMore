# Spliting the dataset into X and y and applying train_test_split

```python
# Assuming that your feature columns are 'sepal_length', 'sepal_width', 'petal_length', and 'petal_width'
X_columns = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
X = X_columns
# Assuming that your target column is 'species'
y = df[['species']]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

# Applying Decision Tree classifier

```python
# Create a Decision Tree classifier
clf = DecisionTreeClassifier(max_depth=3)  # Limit the maximum depth
clf.fit(X_train, y_train)
```

```python
# Visualize the Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(clf, feature_names=X.columns, class_names=clf.classes_, filled=True)
plt.show()
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
# Make predictions on the test data
y_pred = clf.predict(X_test)

print(classification_report(y_test,y_pred))
```

```python
# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
# Display the confusion matrix
print("Confusion Matrix:")
print(conf_matrix)
```
