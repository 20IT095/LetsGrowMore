```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the image
image_path = 'wolf.jpeg'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB
```

```python
# Display the image using matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Turn off axes
plt.title('RGB Image')
plt.show()
```

```python
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)  # Convert to grayscale
```

```python
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
```

```python
image_inverted = 255 - image_gray  # Invert the grayscale image
```

```python
# Display the image using matplotlib
plt.imshow(image_inverted, cmap='gray')
plt.title('Inverted Image')
plt.axis('off')
```

```python
ksize = (21, 21)  # Adjust kernel size based on image size
image_blurred = cv2.GaussianBlur(image_inverted, ksize, sigmaX=0, sigmaY=0)
```

```python
# Display the image using matplotlib
plt.imshow(image_blurred, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')
```

```python
pencil_sketch = cv2.divide(image_gray, image_blurred, scale=256.0)
```

```python
# Display the image using matplotlib
plt.imshow(pencil_sketch, cmap='gray')
plt.title('Pencil Sketch')
plt.axis('off')
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.imshow(image_rgb)
plt.title('Original RGB Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(image_gray, cmap='gray')
plt.title('Original Grey Image')
plt.axis('off')

plt.subplot(133)
plt.imshow(pencil_sketch, cmap='gray')
plt.title('Inverted Blurred Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.imshow(image_blurred, cmap='gray')
plt.title('Blurred Image')
plt.axis('off')

plt.subplot(132)
plt.imshow(pencil_sketch, cmap='gray')
plt.title('Pencil Sketch Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
